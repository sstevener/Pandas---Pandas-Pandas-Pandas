```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
csv_path = "Resources/schools_complete.csv"
csv_path = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
#school_data_df = pd.read_csv(school_data_to_load)
#student_data_df = pd.read_csv(student_data_to_load)
school_data_df = pd.read_csv("Resources/schools_complete.csv")
student_data_df = pd.read_csv("Resources/students_complete.csv") 

school_data_df.head(25)
student_data_df.head(25)

# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])

combined_df = school_data_complete_df.rename(columns={"student_name" : "Student Name", "gender" : "Gender", "grade" : "Grade", "school_name" : "School Name", "reading_score" : "Reading Score", "math_score" : "Math Score", "type" : "School Type", "size" : "School Size", "budget" : "Budget"
})
```

```python
#District Summary

Total_Schools = combined_df["School Name"].nunique()
Total_Students = combined_df["Student Name"].count()
Total_Budget = school_data_df["budget"].sum()
Average_Math = combined_df["Math Score"].mean()
Average_Reading = combined_df["Reading Score"].mean()
Percent_Passing_Math = (combined_df[combined_df["Math Score"] > 70]["Math Score"].count())/(combined_df["Math Score"].count())*100
Percent_Passing_Reading = (combined_df[combined_df["Reading Score"] > 70]["Reading Score"].count())/(combined_df["Reading Score"].count())*100
Percent_Overall_Passing_Rate = (Percent_Passing_Math + Percent_Passing_Reading)/2

District_Summary_data = pd.DataFrame(
    {"Total Schools": [Total_Schools],
     "Total Students": [Total_Students],
     "Total Budget": [Total_Budget],
     "Average Math": [Average_Math],
     "Average Reading": [Average_Reading],
     "% Passing Math": [Percent_Passing_Math],
     "% Passing Reading": [Percent_Passing_Reading],
     "% Overall Passing Rate": [Percent_Overall_Passing_Rate]})  

District_Summary = District_Summary_data[["Total Schools","Total Students", "Total Budget", "Average Math", "Average Reading", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]


Formatted_District_Summary = District_Summary.style.format({"Total Schools":"{:,}",
                                                            "Total Students":"{:,}",
                                                            "Total Budget":"${:,.2f}",
                                                            "Average Math":"{:.2f}", 
                                                            "Average Reading":"{:.2f}",
                                                            "% Passing Math":"{:.2f}%", 
                                                            "% Passing Reading":"{:.2f}%", 
                                                            "% Overall Passing Rate":"{:.2f}%"})

Formatted_District_Summary
```

```python
### School Summary

School_Name = combined_df.set_index("School Name").groupby(["School Name"])

School_Type = combined_df.groupby("School Name")["School Type"]

Total_Students = combined_df.groupby('School Name')["School Size"].count()

Total_School_Budget = combined_df.groupby('School Name')['Budget'].sum()

Per_Student_Budget = (combined_df.groupby("School Name")["Budget"].sum())/(combined_df.groupby("School Name")["School Size"].sum())

Average_Math_Score = combined_df.groupby('School Name')["Math Score"].mean()

Average_Reading_Score = combined_df.groupby('School Name')["Reading Score"].mean()

Pct_Passing_Math = (combined_df[combined_df['Math Score'] >= 70].groupby('School Name')['Student ID'].count()/Total_Students)*100

Pct_Passing_Reading = (combined_df[combined_df['Reading Score'] >= 70].groupby('School Name')['Student ID'].count()/Total_Students)*100 

Pct_Overall_Passing = (combined_df[(combined_df['Reading Score'] >= 70) & (combined_df['Math Score'] >= 70)].groupby('School Name')['Student ID'].count()/Total_Students)*100 

School_Summary_All = pd.DataFrame(
    {"School Name": School_Name,
    "School Type": School_Type,
    "Total Students": Total_Students,
    "Total School Budget": Total_School_Budget,
    "Per Student Budget": Per_Student_Budget,
    "Average Math Score": Average_Math_Score,
    "Average Reading Score": Average_Reading_Score,
    "% Passing Math": Pct_Passing_Math,
    "% Passing Reading": Pct_Passing_Reading,
    "% Overall Passing": Pct_Overall_Passing
    })

Summary_Per_School = School_Summary_All[["School Type",
                                         "Total Students",
                                         "Total School Budget",
                                         "Per Student Budget",
                                         "Average Math Score",
                                         "Average Reading Score",
                                         "% Passing Math",
                                         "% Passing Reading",
                                         "% Overall Passing"]]

#Summary_Per_School = Summary_Per_School.dropna().set_index("School Name")

Summary_Per_School = Summary_Per_School.style.format({"Total School Budget":"${:,.2f}",
                                                          "Per Student Budget":"${:,.2f}",
                                                          "Average Math Score":"{:.2f}",
                                                          "Average Reading Score":"{:.2f}",
                                                          "% Passing Math":"{:.2f}%",
                                                          "% Passing Reading":"{:.2f}%",
                                                          "% Overall Passing":"{:.2f}%"
                                                               })

Summary_Per_School
```

```python
#Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

students9  = student_data_df.loc[student_data_df['grade'] == '9th',]
students10 = student_data_df.loc[student_data_df['grade'] == '10th',]
students11 = student_data_df.loc[student_data_df['grade'] == '11th',]
students12 = student_data_df.loc[student_data_df['grade'] == '12th',]

students9_schoolcount  = students9.groupby('school_name')['grade'].count()
students10_schoolcount = students10.groupby('school_name')['grade'].count()
students11_schoolcount = students11.groupby('school_name')['grade'].count()
students12_schoolcount = students12.groupby('school_name')['grade'].count()

math9_school  = students9.groupby('school_name')['math_score'].sum() / students9_schoolcount
math10_school = students10.groupby('school_name')['math_score'].sum() / students10_schoolcount
math11_school = students11.groupby('school_name')['math_score'].sum() / students11_schoolcount
math12_school = students12.groupby('school_name')['math_score'].sum() / students12_schoolcount

mathscore_by_grade_school = pd.concat([math9_school, 
                                    math10_school,
                                    math11_school,
                                    math12_school],
                                    axis = 1)

mathscore_by_grade_school = mathscore_by_grade_school.rename(columns={0 :'9th', 1 : '10th', 2 : '11th', 3 : '12th'})

formatted_mathscore_by_grade_school = mathscore_by_grade_school.style.format({"math9_school" : "{:.1f}",
                                                          "math10_school" : "{:.1f}",
                                                          "math11_school" : "{:.1f}",
                                                          "math12_school" : "{:.1f}",})

formatted_mathscore_by_grade_school.index.name = "School"

formatted_mathscore_by_grade_school
```

```python
#Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
 
bins = [0,600,626,641,1000]
group_name = ["<$600","$600-625","$626-640",">$640"]
combined_df['spending_bins'] = pd.cut(combined_df['Budget']/combined_df['School Size'], bins, labels = group_name)

by_spending = combined_df.groupby('spending_bins')

avg_math = by_spending['Math Score'].mean()
avg_read = by_spending['Reading Score'].mean()
pass_math = combined_df[combined_df['Math Score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
pass_read = combined_df[combined_df['Reading Score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
overall = combined_df[(combined_df['Reading Score'] >= 70) & (combined_df['Math Score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()

summary_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall
    })
            
summary_by_spend = scores_by_spend[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
    ]]

summary_by_spend.index.name = "Per Student Budget"
scores_by_spend = scores_by_spend.reindex(group_name)

summary_by_spend.style.format({'Average Math Score': '{:.1f}', 
                              'Average Reading Score': '{:.1f}', 
                              '% Passing Math': '{:.1%}', 
                              '% Passing Reading':'{:.1%}', 
                              'Overall Passing Rate': '{:.1%}'})
```
